```python
def create_lags(dataset, n_in=1, n_out=1, dropnan=True):
    """
    Function to create lag variables
    
    Inputs:
        dataset (df): input data
        n_in: Number of lag observations as input(X).
        n_out: Number of observations as output (Y).
        dropnan: Boolean whether or not to drop rows with NaN values.
        
    Output:
        Pandas DataFrame with lag variables for supervised learning
    """
    df = pd.DataFrame(dataset[:])
    
    # input (t-n,...,t)
    for i in range(n_in, 0, -1):
        df['t-'+str(i)] = df[x+'Occ'].shift(i)
    
    df['t'] = df[x+'Occ'].values
    
    # output (t+1, ..., t+n)
    for i in range(1, n_out+1):
        df['t+'+str(i)] = df[x+'Occ'].shift(-i)
        
    # drop rows with NaN values
    if dropnan:
        df.dropna(inplace=True)
        
    df = df.drop(x+'Occ', axis=1)
    
    return df
```

```python
# lag von 5 und +144 Zeitschritte in die Zukunft = 24 Stunden
lagged_data = create_lags(data, 5, 144)
pd.DataFrame.to_csv(lagged_data, 'data_'+x+'_input_5_output_144.txt', sep=',', index=True)
pd.DataFrame.to_csv(lagged_data, 'data_'+x+'_input_5_output_144.csv', sep=',', index=True)
```
