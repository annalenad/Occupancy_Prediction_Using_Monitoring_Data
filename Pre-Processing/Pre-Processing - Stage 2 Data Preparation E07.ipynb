## Data Preparation

```python
# Drop columns for ElPIT and ElPo because of correlation
data = data.drop(x+'ElPIT', axis = 1)
data = data.drop(x+'ElPo', axis = 1)
```

```python
# Drop column with year because it should not be used for training the model
data = data.drop('Year', axis = 1)
```

### One-Hot Encoding

```python
# One-hot encode the data using pandas get_dummies
weekday_name = pd.get_dummies(data['Weekday Name'], prefix='Weekday Name')
day = pd.get_dummies(data['Day'], prefix='Day')
month = pd.get_dummies(data['Month'], prefix='Month')
hour = pd.get_dummies(data['Hour'], prefix='Hour')
minute = pd.get_dummies(data['Minute'], prefix='Minute')
second = pd.get_dummies(data['Second'], prefix='Second')

# Now the columns are transformed into numbers
```

```python
# Concat new columns to the dataset
data = pd.concat([data, month, day, hour, minute, second, weekday_name], axis=1)
```

```python
# save dataframe to csv
pd.DataFrame.to_csv(data, 'data_'+x+'_one-hot-encoded.txt', sep=',', index=True)
pd.DataFrame.to_csv(data, 'data_'+x+'_one-hot-encoded.csv', sep=',', index=True)
```
